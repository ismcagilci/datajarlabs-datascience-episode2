## Bu dersimizde görmediğimiz fakat çok yaygın olarak kullanılan çizge türlerinden birisi de çubuk grafikleri (bar plot)'dir. Çubuk grafikleri özellikle, kategorik (yani sınırlı sayıda farklı değer içeren) değişkenleri incelemek için yararlı araçlardır. Bu ödevde, kategorik bir değişken içeren bir veri kümesi bulmanız gerekecek. Bunun için Kaggle sitesindeki veri setlerine bir göz atabilirsiniz. Daha sonra, bulduğunuz veri kümesini Pandas'ı kullanarak bir veri çerçevesine aktarın. Veri kümenizde yer alan kategorik değişkenlerden iki tanesini seçin ve subplot kullanarak, bu iki değişkenin çubuk grafiklerini çıkarın. Bulgularınızı mentörünüzle tartışmayı unutmayın!

```python
df = pd.read_csv("happiness.csv")
```

```python
df
```

```python
plt.figure(figsize=(10,5))
```

```python
plt.subplot(1,2,1)
plt.plot(df['Ladder score'],df["Freedom to make life choices"],color='purple')
plt.ylabel('Freedom')
plt.xlabel("Happiness")
plt.title('Lineer')
plt.show

plt.subplot(1,2,2)
plt.plot(df['Ladder score'],df["Logged GDP per capita"],color='orange')
plt.ylabel('GDP per capita')
plt.xlabel("Happiness")
plt.title('Lineer')

plt.show()
```
